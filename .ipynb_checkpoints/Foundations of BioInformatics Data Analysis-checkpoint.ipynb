# Fasta format

FASTA is a ubiquitous text-based format for representing nucleotide sequences or amino acid sequences. A FASTA file can contain one sequence or multiple sequences. If a FASTA file contains multiple sequences, it may sometimes be referred to as a “multi-FASTA” file.

Each FASTA entry begins with a > (greater-than) symbol, followed by a comment on the same line describing the sequence that will follow. The actual sequence begins on the line after this comment. Another > (greater-than) symbol denotes the beginning of another FASTA entry (comment describing the sequence + the sequence itself).

EXAMPLE FASTA FILE
This example FASTA file contains two linear nucleotide sequences.
```
>gi|1817694395|ref|NZ_JAAGMU010000151.1| Streptomyces sp. SID7958 contig-52000002, whole genome shotgun sequence
CCGGCTGGCGCGGCTGGCGCTGGCGGTGGGGCTGCGGCTGCTGGAGCTGGGGGTGGCGCTGGAGGCGCAC
GGCCAGAACCTGCTGGTGGTGCTGTCGCCGTCCGGGGAGCCGCGGCGGCTGGTCTACCGCGATCTGGCGG
ACATCCGGGTCTCCCCCGCGCGGCTGGCCCGGCACGGTATCCGGGTTCCGGACCTGCCGGCG

>gi|1643051563|gb|SZWM01000399.1| Citrobacter sp. TBCS-14 contig3128, whole genome shotgun sequence
GCACAGTGAGATCAGCATTCCGTTGGATCTACTGGTCAATCAAAACCTGACGCTGGGTACTGAATGGAAC
CAGCAGCGCATGAAGGACATGCTGTCTAACTCGCAGACCTTTATGGGCGGTAATATTCCAGGCTACAGCA
GCACCGATCGCAGCCCATATTCGAAAGCCGAGATCTTCTCTTTGTTTGCCGAAAACAACATG
```

note here that a multi-line fasta sequence is 80 characters or less (often 70). 

## So why are we concerned about FASTA

Most (if not all) reference genomes are in FASTA format. 

There are many resources for downloading reference genome sequences, with some subtle differences between them. A few common ones are 

- UCSC: https://genome.ucsc.edu/cgi-bin/hgGateway
- Ensembl: https://useast.ensembl.org/index.html
- RefSeq: https://www.ncbi.nlm.nih.gov/refseq

### Caveats

There can be different naming conventions for chromosomes - 
- deflines from a UCSC human genome looks like this:
```
>chr1
>chr2
>chr3
>chr4
>chr5
```

Compared to Ensembl human genome deflines, which looks like this:

```
>1 dna:chromosome chromosome:GRCh38:1:1:248956422:1 REF
>2 dna:chromosome chromosome:GRCh38:2:1:242193529:1 REF
>3 dna:chromosome chromosome:GRCh38:3:1:198295559:1 REF
>4 dna:chromosome chromosome:GRCh38:4:1:190214555:1 REF
>5 dna:chromosome chromosome:GRCh38:5:1:181538259:1 REF
```

There are also differences in gene models – one example is PIK3CA (https://doi.org/10.1186/s12864-015-1308-8), where the Ensembl annotation is longer than the RefGene version.
**Whichever reference genome you pick, be consistent (e.g., use a properly corresponding annotation file).**

You might also be asked for `FASTA` or `FASTQ` indices or `.fai` files (http://www.htslib.org/doc/faidx.html).
- These are tab-delimited text files and allow fast searching of the files.

There are 5 tab-delimited columns for FASTA indices and 6 tab-delimited columns for FASTQ indices:
- Name: Name of the reference sequence
- Length: Total length of the reference sequence
- Offset: Offset of the sequence’s first base (the position of the sequence in the fasta in terms of character count -- including newline characters, which means the newline encoding will matter here).
- Linebases: Number of bases on each line
- Linewidth: Number of bytes or characters in each line including the newline character.
- Qualoffset: Offset of sequence’s first quality within the FASTQ file (FASTQ only).

## Sequence Data

Data in the SRA, European Bioinformatics Institute (EBI), and DNA Database of Japan (DDBJ) are all shared across all three resources.

### NCBI Short Read Archive (SRA)

The Sequence Read Archive (SRA) is the NIH’s repository of high-throughput sequencing data, including FASTQ files. 

The SRA is organized into four hierarchical levels: 
- Study (SRP/ERP/DRP) – overarching project
- Sample (SRS/ERS/DRS) – biological samples (mouse or human cell cultures)
- Experiment (SRX/ERX/DRX) – experiments on each sample (three treatment and three control samples)
- Run (SRR/ERR/DRR) – sequencing run per experiment (might be multiple runs per experiment if it was split across multiple lanes).

 ![SRA structure](imgs/SRA_structure.png)
 
 Downloading the data in a usable format usually involves using the sra-toolkit https://trace.ncbi.nlm.nih.gov/Traces/sra/sra.cgi?view=software
 
### European Nucleotide Archive ENA

You can also download reads from ENA using `wget` by searching for the Accession in the box labeled ‘Enter accession’ here: https://www.ebi.ac.uk/ena/browser/home 


# FASTQ
The FASTQ format is an extension of FASTA that stores both biological sequences (usually nucleotide sequences) and their corresponding quality scores. Both the sequence letter and quality score are encoded with a single character for brevity.

A FASTQ file normally uses four lines per sequence:

- A line beginning with @ followed by a sequence identifier and optional description (like the comment line at in a FASTA file)
- The raw sequence letters
- A line beginning with +, sometimes followed by the same comment as the first line
- A line encoding the quality values for the sequence in line 2, with the same numbers of symbols as letters in the sequence



EXAMPLE FASTQ FILE

```
@SRR8933535.1 1 length=75
NAGGAAACAAAGGCTTACCCGTTATCATTTCCGCAAGAATGCACCCACACGACCATATATCAATGGATGTGGAGT
+SRR8933535.1 1 length=75
#AAAAEEEEEEEEEEEEEEEEEEEEEAEEEEEAEEEEEEEAEAEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEAE

@SRR8933535.2 2 length=75
NGAGGAGTGGTGGTAGTGTTGCTTGGTGGCAAAGATGTAGTTGGTGGGAAAGCTGAAGTGGTACCGTTGGTTGGA
+SRR8933535.2 2 length=75
#AAAAEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEAAEEEE
```

Basically Fastq files are text files. 

### Why can't I open my fastq file in a text editor ?

Most fastq files are compressed with Gzip so you will usually see the file extension `.gz` . This can result in significant space savings especially given that sequence data is highly repetitive compared to normal text. 

### How then do i generally view them?

Use either `zcat < yourfastq.fq.gz | head -8` or `gunzip -c yourfastq.fq | head -8 ` to view the first two reads of your own FASTQ (you should probably just leave them all compressed as .gz files since they take up less space). 


### Single End vs Paired End Reads


![Read types](imgs/pe_reads_img.png)

- If your reads are single-end, each library will have just one read.
- If your reads are paired-end, each library will (usually) have two reads.

The paired-end read files should both have the same number of reads and be sorted in the same order based on their read names to maintain pair concordance.
Be careful if trying to manually parse or filter your paired-end reads -- you want to maintain that pair concordance.

# Alignment formats
These are formats for storing alignments of nucleotide or amino acid sequences. The alignment formats discussed here can be manipulated using SAMtools.

## SAM
Sequence Alignment/Map (SAM) is a text-based alignment format that supports single- and paired-end reads produced by different sequencing platforms. It can support short and long reads (up to 128Mbp). The format has been extended to include unmapped sequences, and it may contain other data such as base-call & alignment qualities. (https://samtools.github.io/hts-specs/SAMv1.pdf)

The SAM format consists of a header and an alignment section. Headings begin with the @ symbol, which distinguishes them from the alignment section. All lines in a SAM file are tab-delimited. Alignment sections contain 11 mandatory fields, with other fields being optional. Although the mandatory fields must be present, their value can be a * or a 0 depending on the field. Optional fields are presented as key-value pairs in the format TAG:TYPE:VALUE.

### SAM HEADER

Each line in the header section starts with an `@` symbol and a two-letter record type code.
`@HD` line contains metadata and must be single first line if present 
`@PG` contains program-level information
`@SQ` contains information about the reference sequence
`@RG` is contains information about the different read groups (absent in the following example).

Example header of a SAM file

```
@HD    VN:1.0    SO:unsorted
@PG    ID:GSNAP    PN:gsnap    VN:2017-01-14    CL:/gpfs/runtime/cbc_conda/cbc_conda_v1_root/envs/cbc_conda_v1/bin/gsnap.nosimd -t 8 --gunzip -A sam -N1 --use-shared-memory=0 -d Ensembl_mus_GRCm38.p5_rel89 -s /gpfs/data/cbc/cbcollab/cbc_ref/gmapdb_2017.01.14/Ensembl_mus_GRCm38.p5_rel89/Ensembl_mus_GRCm38.p5_rel89.maps/Mus_musculus.GRCm38.89.splicesites.iit /gpfs/data/cbc/Muscle_RNA-seq_Jaime/fastq/SRS643406_SRX612440.fq.gz
@SQ    SN:1    LN:195471971
@SQ    SN:2    LN:182113224
@SQ    SN:3    LN:160039680
@SQ    SN:4    LN:156508116
@SQ    SN:5    LN:151834684
@SQ    SN:6    LN:149736546
```

### Alignment Section

The mandatory fields in a SAM file are…

- QNAME Query template name. Reads/segments with the same QNAME are from the same template. A read may occupy multiple alignment lines when its alignment is chimeric or when multiple mappings are given.
- FLAG Bitwise flag (pairing, strand, mate strand, etc.).
- RNAME Reference sequence name.
- POS 1-based leftmost mapping position. The first base in a reference sequence has coordinate 1. POS is set to 0 for an unmapped read.
- MAPQ Mapping quality. If equal to 255, the mapping quality is not available.
- CIGAR Extended Consise Idiosyncratic Gapped Alignment Report string. M for match/mismatch, I for insertion and D for deletion compared with the reference, N for skipped bases on the reference, S for soft clipping, H for hard clipping, and P for padding.
- RNEXT Reference name of the mate/next read in the template. For the last read, the next read is the first read in the template.
- PNEXT Position of the primary alignment of the mate/next read.
- TLEN Observed template length.
- SEQ Segment sequence.
- QUAL Phred-based base quality (same as the quailty string in FASTQ) plus 33.

A detailed specification is provided here https://samtools.github.io/hts-specs/SAMv1.pdf

## BAM and CRAM

SAM files can get very large but they can be converted into BAM files, which are just block compressed binary versions of SAM files that take up less space and allows quick access to a given alignment.
- Like SAM files, they contain a header section and an alignment section, you can view the contents of a BAM file using `samtools view` (http://www.htslib.org/doc/samtools-view.html). 
- BAM files can be quickly queried if they are also indexed, which organizes your alignments into bins and is like making a table of contents for your alignments.
- BAM indexes are stored in BAI files, use `samtools index` for this (http://www.htslib.org/doc/samtools-index.html).
- A BAM file must be sorted before indexing, use `samtools sort` for this (http://www.htslib.org/doc/samtools-sort.html).
- By default, `samtools sort` will sort based on chromosomal coordinates, but the `n` flag can be specified if read name sorting is required.

# BED format

BED files are tab or space-delimited text files to store genomic regions and annotations and are mainly used as a way to define which lines are shown in on an annotation track (can use with UCSC genome browser or IGV).

There are only three required columns:
- chrom: chromosome or scaffold name
- chromStart: start site on chromosome (starts at 0)
- chromEnd: end site on chromosome (up to this site will be shown -- chromStart=0 and chromEnd=100 will span bases 0-99 (but not 100). 

There’s also 9 additional optional fields:
- name: name of the BED line.
- score: A score between 0 and 1000 (UCSC browser will shade higher colors darker shades of grey).
- strand: Strand orientation (-, +, or . if null/no strand)
- thickStart: start coordinate where the annotation will be displayed in a thicker way (e.g., start codon)
- thickEnd: stop coordinate where the annotation will stop being displayed in a thicker way (e.g., stop codon).
- itemRgb: (R,G,B) color values determining the color that should be displayed for the annotation.
- blockCount: number of blocks (exons) on the line of the BED
- blockSizes: comma-separated values for the sizes of the blocks in blockCount
- blockStarts: comma-separated values for the coordinates where the blocks start.


Can be used for manipulating and the so called 'genome arithmetic' using the powerful bedtools suite
https://bedtools.readthedocs.io/en/latest/

# GTF and GFF3 

Reference genomes will have annotations associated with them, usually `.gtf` or `.gff3`, which are text files that describe different genomic features. 
Each of these files contain 9 tab-separated fields, none of which can be empty (but could have a `.` to indicate a null value instead of an empty field): 

- seqname - name of the chromosome or scaffold; chromosome names can be given with or without the 'chr' prefix.
- source - name of the program that generated this feature, or the data source (database or project name)
- feature - feature type, must be either 
- start - Start position of the feature, with sequence numbering starting at 1.
- end - End position of the feature, with sequence numbering starting at 1.
- score - A floating point value.
- strand - defined as + (forward) or - (reverse).
- frame or phase - One of '0', '1' or '2'. '0' indicates that the first base of the feature is the first base of a codon, '1' that the second base is the first base of a codon, and so on..
- attribute - A semicolon-separated list of tag-value pairs, providing additional information about each feature.

The main distinction is that `.gff3` files allow for 3 levels of hierarchical groupings of features and sub-features (i.e., genes, transcripts, exons), and allows each sub-feature to belong to more than one feature (i.e., an exon can be used in more than one gene or transcript).

Note that .gff files are not the same as .gff3 files.

# VCF files

VCF, or Variant Call Format, It is a standardized text file format used for representing SNP, indel, and structural variation calls. 

### Structure of a VCF file
A valid VCF file is composed of two main parts: the header, and the variant call records.

![vcf file](imgs/vcf_structure.png)

The header contains information about the dataset and relevant reference sources (e.g. the organism, genome build version etc.), as well as definitions of all the annotations used to qualify and quantify the properties of the variant calls contained in the VCF file. The header of VCFs generated by GATK tools also include the command line that was used to generate them. Some other programs also record the command line in the VCF header, but not all do so as it is not required by the VCF specification. For more information about the header, see the next section.

The actual data lines will look something like this:
